```python
# Implementing Forward Propagation Algorithm
import numpy as np
input_data = np.array([2,3])

# Defifning weights as a dictionary
weights = {'node_0':np.array([1,1]),
           'node_1':np.array([-1,1]),
           'output':np.array([2,-1])}
# Define the value of the nodes in the hidden layer
node_0_value = (input_data*weights['node_0']).sum()
node_1_Value = (input_data*weights['node_1']).sum()

# Store the values of hidden layer in an array
hidden_layer = np.array([node_0_value, node_1_Value])

# Caluclate the output
output = (hidden_layer*weights['output']).sum()
```

**Example 02:**
<p style="color:red;"> Activation Function: tanh

```python
# Implementing Forward Propagation Algorithm
import numpy as np
input_data = np.array([2,3])

# Defifning weights as a dictionary
weights = {'node_0':np.array([1,1]),
           'node_1':np.array([-1,1]),
           'output':np.array([2,-1])}
# Define the value of the nodes in the hidden layer
node_0_input = (input_data*weights['node_0']).sum()
node_0_output = np.tanh(node_0_input)
node_1_input = (input_data*weights['node_1']).sum()
node_1_output = np.tanh(node_1_input)

# Store the values of hidden layer in an array
hidden_layer = np.array([node_0_output, node_1_output])

# Caluclate the output
output = (hidden_layer*weights['output']).sum()

print(output)
```

**Example 03:**
<p style="color:red;"> Activation Function: ReLU

```python
input_data = np.array([3,5])

# Defifning weights as a dictionary
weights = {'node_0':np.array([2,4]),
           'node_1':np.array([4,-5]),
           'output':np.array([2,7])}

# ------------------------------------------------------------------
def relu(input):
    '''Define your relu activation function here'''
    # Calculate the value for the output of the relu function: output
    output = max(0, input)
    
    # Return the value just calculated
    return(output)
# ------------------------------------------------------------------

# Calculate node 0 value: node_0_output
node_0_input = (input_data * weights['node_0']).sum()
node_0_output = relu(node_0_input)

# Calculate node 1 value: node_1_output
node_1_input = (input_data * weights['node_1']).sum()
node_1_output = relu(node_1_input)

# Put node values into array: hidden_layer_outputs
hidden_layer_outputs = np.array([node_0_output, node_1_output])

# Calculate model output (do not apply relu)
model_output = (hidden_layer_outputs * weights['output']).sum()

# Print model output
print(model_output)
```

**Example 04:**
    <p style="color:red;"> Multiple input prediction

```python
input_data = [np.array([3,5]), np.array([1,-1]), np.array([0,0]), np.array([8,4])]

# Defifning weights as a dictionary
weights = {'node_0':np.array([2,4]),
           'node_1':np.array([4,-5]),
           'output':np.array([2,7])}

def predict_with_network(input_arg, weights):
    
    node_0_input = (input_arg * weights['node_0']).sum()
    node_0_output = relu(node_0_input)

    # Calculate node 1 value: node_1_output
    node_1_input = (input_arg * weights['node_1']).sum()
    node_1_output = relu(node_1_input)

    # Put node values into array: hidden_layer_outputs
    hidden_layer_outputs = np.array([node_0_output, node_1_output])

    # Calculate model output (do not apply relu)
    last_layer_input = (hidden_layer_outputs * weights['output']).sum()
    model_output = relu(last_layer_input)
    
    return(model_output)

results = []

for row in input_data:
    results.append(predict_with_network(row, weights))


print(results)
```

```python
input_data = np.array([3,5])

# Defifning weights as a dictionary
weights = {'node_0_0':np.array([2,4]),
           'node_0_1':np.array([4,-5]),
           'node_1_0':np.array([-1,2]),
           'node_1_1':np.array([1,2]),
           'output':np.array([2,7])}


def predict_with_network(input_data):
    # Calculate node 0 in the first hidden layer
    node_0_0_input = (input_data * weights['node_0_0']).sum()
    node_0_0_output = relu(node_0_0_input)

    # Calculate node 1 in the first hidden layer
    node_0_1_input = (input_data * weights['node_0_1']).sum()
    node_0_1_output = relu(node_0_1_input)

    # Put node values into array: hidden_0_outputs
    hidden_0_outputs = np.array([node_0_0_output, node_0_1_output])
    
    # Calculate node 0 in the second hidden layer
    node_1_0_input = (hidden_0_outputs * weights['node_1_0']).sum()
    node_1_0_output = relu(node_1_0_input)

    # Calculate node 1 in the second hidden layer
    node_1_1_input = (hidden_0_outputs * weights['node_1_1']).sum()
    node_1_1_output = relu(node_1_1_input)

    # Put node values into array: hidden_1_outputs
    hidden_1_outputs = np.array([node_1_0_output, node_1_1_output])

    # Calculate model output: model_output
    model_output = (hidden_1_outputs * weights['output']).sum()
    
    # Return model_output
    return(model_output)

output = predict_with_network(input_data)
print(output)
```
