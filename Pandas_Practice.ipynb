## Building Pandas Dataframe from scratch

**Using List**

```python
names = ['Shah', 'Ahmed', 'Aghil', 'Abbas', 'Hossein' ]
Age = [23, 45, 34, 13, 25]
City = ['Shiraz', 'Haifa', 'Tadmor', 'Goldcoast', 'Tubingen']
wage_per_hour = [21, 23, 56, 12, 10]

column_names = ['Name', 'Age', 'City', 'Salary']
list_of_data = [names, Age, City, wage_per_hour]

zipped_List = list(zip(column_names, list_of_data))
dict_data = dict(zipped_List)

df = pd.DataFrame(dict_data)
df.head()
```

**Using dict**

```python
data_as_dict = {'Name': ['Shah', 'Ahmed', 'Aghil', 'Abbas', 'Hossein' ],
               'Age': [23, 45, 34, 13, 25], 
               'City': ['Shiraz', 'Haifa', 'Tadmor', 'Goldcoast', 'Tubingen'],
               'Salary':[21, 23, 56, 12, 10]}
data_as_pd_df = pd.DataFrame(data_as_dict)
data_as_pd_df.head()
```

```python
data_as_pd_df.index.name = 'row'
data_as_pd_df.head()
```
